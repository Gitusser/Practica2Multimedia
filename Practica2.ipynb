```python
path = r'/multimedia/imagenesoriginales/000.png'
x = cv2.imread(path)
plt.imshow(x, cmap="gray")
```

```python
listaMSEx=[]
listaMSEy=[]
listaSNRx=[]
listaSNRy=[]
listaPSNRy=[]
listaSSIMy=[]
listaPCCy=[]
listaRMSEy=[]
```

```python
for i in [5,35,65,95,128,155,185,210]:
    pathc = "/multimedia/imagenescuantificadas"+str(i)+"/000.png"
    y = cv2.imread(pathc)
    MSE = skimage.metrics.mean_squared_error(x,y)
    listaMSEy.append(MSE)
    sizefile = os.path.getsize(pathc)
    listaMSEx.append(sizefile)
    
    SNR = skimage.metrics.peak_signal_noise_ratio(x, y, data_range=np.std(x))
    listaSNRy.append(SNR)
    
    PSNR = skimage.metrics.peak_signal_noise_ratio(x, y, data_range=255)
    listaPSNRy.append(PSNR)
    PCC = scipy.stats.pearsonr(x.flatten(),y.flatten())[0]
    listaPCCy.append(PCC)
    RMSE = math.sqrt(MSE)
    listaRMSEy.append(RMSE)
```

```python
def graficar(lx,f1,f2,f3, titulo):
    plt.figure()  # Comenzamos un nuevo gráfico (figura)
    plt.title(titulo)
    plt.xlabel("abscisa")
    plt.ylabel("ordenada")
    #indice = np.arange(8)   # Declara un array
    #plt.xticks(indice, ("A", "B", "C", "D", "E", "F", "G", "H"))  
    plt.plot(lx, f1, 'b.', lx, f2, 'rd', lx, f3, 'g^')
    #plt.yticks(np.arange(0,100,100))
    
def graficar1(lx,f1,titulo):
    plt.figure()  # Comenzamos un nuevo gráfico (figura)
    plt.title(titulo)
    plt.xlabel("abscisa")
    plt.ylabel("ordenada")
    indice = np.arange(1000)   # Declara un array
    #plt.xticks(indice, ("Q1", "B", "C", "D", "E", "F", "G", "H"))  
    plt.plot(lx, f1, 'b.')
    #plt.yticks(np.arange(0,1,100))
graficar1(listaMSEx,listaMSEy,"MSE")  

```
